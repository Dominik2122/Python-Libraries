```python
import requests, re
from bs4 import BeautifulSoup

r = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c=r.content

soup = BeautifulSoup(c,'html.parser')
all = soup.find_all('div', {"class":"propertyRow"})
i = 0
l = []

for item in all:
    d = {}
    price = all[i].find_all('h4')[0].text
    d['price'] = " ".join(price.split())
    j = 0
    list_of_address = all[i].find_all('span', {'class':'propAddressCollapse'})
    address = []
    for line in list_of_address:
        line_of_address = all[i].find_all('span', {'class':'propAddressCollapse'})[j].text
        address.append(line_of_address)
        j +=1
    d['address'] = address[0] + ', ' + address[1]
    try:
        d['beds'] = all[i].find_all('span', {'class':'infoBed'})[0].text
    except:
        pass
    i += 1
    l.append(d)
    
 
l
```

```python
import pandas as pd
data = pd.DataFrame(l)
data
```
